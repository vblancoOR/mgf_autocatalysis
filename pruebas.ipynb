```python
def powerset(Sm, Sp):
    PS=[]
    N, M = Sm.shape
    Set = list(range(N))
    for i in range(1 << N):
        ss = [Set[j] for j in range(N) if (i & (1 << j))]
        if len(ss)>=2:
            isA_sp, isA_re = IsAutonomous(Sm, Sp, ss, range(M))
            if isA_sp and isA_re:
                PS.append(ss)

    return PS
```
